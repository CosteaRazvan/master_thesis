```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
csv_data = pd.read_csv("fmri_full_data_11_23_2024.csv")
```

```python
csv_data
```

```python
csv_data["Group"].unique()
```

```python
group_counts = csv_data["Group"].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(group_counts.index, group_counts.values, color='skyblue', edgecolor='black')
plt.xlabel('Group Categories')
plt.ylabel('Frequency')
plt.title('Histogram of Group Categories')
plt.xticks(rotation=45)
plt.tight_layout()
```
